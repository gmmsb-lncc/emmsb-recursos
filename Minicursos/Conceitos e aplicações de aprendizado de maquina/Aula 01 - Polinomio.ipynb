```python
from sklearn import datasets
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-1, 1, 10)
```

```python
print(x)
n = len(x)
```

```python
np.random.seed(18)
y = x**3 
y = y * 0.05*np.random.rand(n)
```

```python
y
```

```python
plt.plot(x, y, 'o')
```

```python
t = np.linspace(-1, 1, 200)
p = np.poly1d(np.polyfit(x, y, 1))
plt.plot(x, y, 'o', t, p(t), '-')
print("Errro no treinamento: ", sum(abs(p(x) - y)))
```
